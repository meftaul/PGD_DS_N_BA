# Data Collection

```python
url = 'https://www.thedailystar.net/top-news/news/rmg-can-gain-additional-489m-global-online-business-study-3528361'
```

```python
# fetch the article using beautifulsoup
from bs4 import BeautifulSoup
import requests
```

```python
# request the url if it is accessible then parse the html
req = requests.get(url)
print(req.text)
```

```python
soup = BeautifulSoup(req.text, 'html.parser')
raw_article = soup.find('article')
print(raw_article.text)
```

```python
# detect if there are many new lines
import re
new_line_regex = re.compile(r'\n+')
article = new_line_regex.sub('\n', raw_article.text)
print(article)
```

```python
# remove last three lines
article = article.split('\n')
article = article[:-4]
article = '\n'.join(article)
print(article)
```

```python
title = article.split('\n')[2]
```

```python
main_article = '\n'.join(article.split('\n')[6:])
print(main_article)
```

```python
# Remove New Lines
main_article = main_article.replace('\n', '')
print(main_article)
```

## Fetch article using newspaper3k

```python
from newspaper import Article
article = Article(url)
article.download()
article.parse()

print(f'Title: {article.title}')
print(f'Authors: {article.authors}')
print(f'Publication Date: {article.publish_date}')
print(f'Article Text: \n {article.text}')
```

```python
main_article = article.text
main_article
```

# Tokenization, Lemmatization, Stopwords Removal

```python
# count the punctuation marks
import string
punctuations = string.punctuation

print(f'Punctuations: {punctuations}')


# count the number of punctuations
count = 0
for char in main_article:
  if char in punctuations:
    count += 1

print(f'Number of Punctuations: {count}')
```

```python
# Count the number of stopwords use spaCy library
import spacy
nlp = spacy.load('en_core_web_sm')
stopwords = spacy.lang.en.stop_words.STOP_WORDS

print(f'Number of Stopwords in spacy: {len(stopwords)}')
print(f'Stopwords: {stopwords}')

# count the number of stopwords
count = 0
for word in main_article.split():
  if word in stopwords:
    count += 1

print(f'Number of Stopwords: {count}')
```

```python
doc = nlp(main_article)
# Tokens and their attributes formatted string
for token in doc:
    # print formatted text top 10 words
    print(f'{token.text:10} {token.pos_:10} {token.lemma_:10} {token.is_stop}')
```

```python
tokens = [token.text for token in doc if token.text not in stopwords and token.text not in punctuations]
tokens
```

```python
# print sentences
sentences = [sent.text for sent in doc.sents]
print(len(sentences))
sentences
```

```python
# print the vocabulary
vocabulary = list(set(tokens))
len(vocabulary)
```

# Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer()
vectorizer.fit_transform(sentences)
sorted_vocab = sorted(vectorizer.vocabulary_.keys())
print(len(sorted_vocab))
sorted_vocab
```

```python
# encode document
vector = vectorizer.transform(sentences)
# summarize encoded vector
print(vector.shape)
print(vector.toarray())
```

# TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
vectorizer.fit(sentences)

sorted_vocab = sorted(vectorizer.vocabulary_)
print(len(sorted_vocab))
sorted_vocab
```

```python
vector = vectorizer.transform(sentences)
print(vector.shape)
print(vector.toarray())
```
