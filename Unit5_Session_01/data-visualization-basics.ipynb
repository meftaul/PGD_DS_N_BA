```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# theme
plt.style.use('ggplot')
```

```python
tips_df  = sns.load_dataset('tips')

type(tips_df)
```

```python
tips_df.head()
```

```python
plt.plot([1,2,3,4])
plt.ylabel('some numbers')
plt.xlabel('some numbers')
plt.title('Plot title')
plt.show()
```

```python
fig, ax = plt.subplots(
    nrows=2,
    ncols=1,
    figsize=(10,8),
)

fig.patch.set_facecolor('lightgrey')
fig.suptitle('Overall Plot Title')

ax[0].hist(tips_df['total_bill'], bins=20, color='lightblue')
ax[0].set_title('Histogram of Total Bill')
ax[0].set_xlabel('Frequency')
ax[0].set_ylabel('Total Bill')

ax[1].hist(tips_df['tip'], bins=20, color='lightgreen')
ax[1].set_title('Histogram of Tips')
ax[1].set_xlabel('Frequency')
ax[1].set_ylabel('Tips')

plt.tight_layout()


plt.show()
```

```python
tips_df.plot.scatter(x='total_bill', y='tip', c='size', colormap='viridis', s=50, alpha=0.6, figsize=(10,8))
plt.show()
```

```python
sns.pairplot(tips_df)
plt.show()
```

```python
plt.subplot(3,1,1)
plt.hist(tips_df['total_bill'], bins=20, color='lightblue')
plt.title('Histogram of Total Bill')

plt.subplot(3,1,2)
plt.hist(tips_df['tip'], bins=20, color='lightgreen')
plt.title('Histogram of Tips')

plt.subplot(3,1,3)
plt.hist(tips_df['size'], bins=20, color='lightcoral')

plt.tight_layout()

plt.show()
```

```python
plt.boxplot(tips_df['total_bill'])
plt.title('Boxplot of Total Bill')
plt.show()
```

```python
tips_df['day'].value_counts()
```

```python
sns.boxplot(x='day', y='total_bill', data=tips_df)
plt.title('Boxplot of Total Bill by Day')
plt.show()
```

```python
plt.boxplot([tips_df[tips_df['day'] == 'Thur']['total_bill'],
             tips_df[tips_df['day'] == 'Fri']['total_bill'],
             tips_df[tips_df['day'] == 'Sat']['total_bill'],
             tips_df[tips_df['day'] == 'Sun']['total_bill']],
            labels=['Thur', 'Fri', 'Sat', 'Sun'])

plt.title('Boxplot of Total Bill by Day')
plt.show()
```
